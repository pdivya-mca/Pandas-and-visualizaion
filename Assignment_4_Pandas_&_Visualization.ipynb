**Department wise toppers (horizantal bar graph or any visual representations of your choice)**

```python
topper= pd.DataFrame(df.groupby('Department')['CodeKata Score'].nlargest(1))
```

```python
topper=topper.droplevel(level=1)
topper
```

```python
topper=topper.sort_values(by='CodeKata Score')
```

```python
topper.plot(kind='barh', color='cyan')
plt.xlabel('CodeKata Score')
```
